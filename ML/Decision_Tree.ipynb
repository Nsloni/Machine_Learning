```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()

# Create a DataFrame from the data
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# The target variable is stored in iris.target
target = iris.target

# Create a copy of the DataFrame
df1 = df.copy()

# The target variable is not a column in the DataFrame, so you don't need to drop it
```

```python
# Define atrribute

x = df1
```

```python
target
```

```python
y = target
```

```python
# Data 80:20 ratio

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
# Decision Tree Algorithm

from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree Classifier
dtree = DecisionTreeClassifier()

# Train the Decision Tree Classifier
dtree.fit(x_train, y_train)

print('Decision Tree Classifier Created')
```

```python
# Predicting Value of test data
# 80 % ----> train_data_ser
# 20 % ----> Store in test_size

from sklearn.metrics import classification_report
y_pred = dtree.predict(x_test)
print("Classification report -\n",classification_report(y_test, y_pred))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

fn = ['sepal length (cm)', 'sepal width (cm)','petal length (cm)','petal width (cm)']
cn = ['sentosa', 'versicolor', 'virginica']

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)

tree.plot_tree(dtree,
              feature_names=fn,
              class_names=cn,
              filled=True);

fig.savefig('Image_Decision_Tree.png')
```
