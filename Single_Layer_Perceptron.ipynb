```python
import pandas as pd

dataset = pd.read_csv("Deep_Learning.csv")
```

```python
dataset.head(3)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'dataset' is your DataFrame
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa", y="score", data = dataset, hue="placed")
plt.show()

# plt.figure(figsize=(4,3))
# sns.scatterplot(x="cgpa",y="acore",data = dataset, hue="placed")
# plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_trains, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron()
pr.fit(x_trains, y_train)
```

```python
pr.score(x_trains, y_train)*100, pr.score(x_test, y_test)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=pr)
plt.show()
```
